```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
polygon_points = pd.read_csv('table1.csv')
lines = pd.read_csv('table2.csv')
```

```python
pol = list(polygon_points.values)
```

```python
pol.append(pol[0])
```

```python
def get_line_coords(line):
    line = line.values
    return [line[0], line[2]], [line[1], line[3]]
```

```python
coords = list(lines.apply(lambda line: get_line_coords(line), axis=1))
```

```python
plt.figure(figsize=(10,10))

for i, line_coords in enumerate(coords):
    plt.plot(line_coords[0], line_coords[1], marker = 'o', linewidth=2, markersize=5)
    #plt.text(line_coords[0][0], line_coords[1][0], f'{line_coords[0][0]}, {line_coords[1][0]}', horizontalalignment='right')
    #plt.text(line_coords[0][1], line_coords[1][1], f'{line_coords[0][1]}, {line_coords[1][1]}', horizontalalignment='right')
xs, ys = zip(*pol)
plt.plot(xs,ys,'D-r', linewidth=5, markersize=10)
plt.show()
```
